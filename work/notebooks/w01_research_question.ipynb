## 2. The question: decision, action, cost of a wrong call

* **What decision does this work improve?**
  It improves the decision of *which page to review and refresh first* in the editorial queue.
  
* **Who acts on it, and what do they do?**
  Content editors and SEO managers act on the output. They will check the highest-ranked pages in the priority queue and perform updates (e.g., rewriting thin sections, expanding content depth, updating obsolete data, or adjusting titles/meta descriptions).
  
* **What does a wrong recommendation cost?**
  * A **False Positive** (recommending a refresh for a page that doesn't need it or has no traffic potential) costs wasted editor hours spent analyzing and rewriting content that yields no return.
  * A **False Negative** (missing a high-impact page that is steadily declining) costs lost search visibility, reduced clicks, and a drop in business traffic/revenue.

```python
# Decision framing summary
print("Decision: Priority ranking for content refresh queue.")
print("Customer: Content editors/SEO managers.")
print("Error costs: Wasted editor time vs. missed traffic/revenue recovery.")
```

## 3. Quick look at the data (2-3 real numbers)

Below, we load the starter CSV (`data/raw/content_refresh_anonymized.csv`) and output key metrics that illustrate the scope and importance of the refresh problem.

```python
import pandas as pd

# Load starter dataset
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

# 1. Overall volume
print(f"Total content items in starter dataset: {len(df):,}")

# 2. Prevalence of traffic decline
decline_pct = (df['trend_direction'] == 'down').mean() * 100
print(f"Percentage of pages currently classified as declining ('down'): {decline_pct:.2f}%")

# 3. Median impressions for pages with active visibility
median_impr = df[df['impressions_90d'] > 0]['impressions_90d'].median()
print(f"Median 90-day impressions for visible pages: {median_impr:,.1f}")
```

## 4. Careful words: what I can and can't claim

* **What we CAN claim:**
  * **Observational association:** We can say that pages with certain attributes (e.g. high age, thin content, low engagement) are historically correlated with traffic decline.
  * **Directional/Decision-Support value:** We can rank pages in terms of relative risk/opportunity to help editors allocate their time more efficiently.
  
* **What we CANNOT claim:**
  * **Causal proof:** We cannot claim that a page's age or thin content *caused* the traffic decline, nor can we guarantee that refreshing it *will cause* traffic to recover.
  * **Reverse-engineering Google:** We are not modeling the search engine algorithm or predicting Google's internal scoring weights; we are simply predicting observed traffic outcomes from historical signals.

```python
# Claim boundaries
print("Claim limits: Decision-support and directional indicators only. No causal or algorithm reverse-engineering claims.")
```
